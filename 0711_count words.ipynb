# Count numbers of words in a text¶

```python
import operator
word_dict={}
text=open("heidi.txt")

line=0
for textline in text:
    line += 1     # line number of the text
    n=0
    for word in textline.strip().split():
        n += 1
        if word in word_dict:
            word_dict[word] += 1
        else:
            word_dict[word] = 1
    #print(f'Number of words in line {line} = {n}')

# sort of word_dict as use of operator values() with decending order
word_sort = sorted(word_dict.items(),key=operator.itemgetter(1),reverse=True)


# words with special character 
nword={}
scword={}
line=0
for chk in word_dict:
   
    if chk.find(".")!= -1:   # word에 . 이 있는가? -1이 아니면 . 있다
        chk=chk.replace(".", "")
        if chk in scword:
            scword[chk] += 1
            line += 1
        else:
            scword[chk] = 1 
            line += 1
    elif chk.find("?") != -1:
        chk=chk.replace("?", "")
        if chk in scword:
            scword[chk] += 1
            line += 1
        else:
            scword[chk] = 1 
            line += 1
        
    elif chk.find("!") != -1:
        chk=chk.replace("!", "")
        if chk in scword:
            scword[chk] += 1
            line += 1
        else:
            scword[chk] = 1 
            line += 1
        
    elif chk.find(",") != -1:
        chk=chk.replace(",", "")
        if chk in scword:
            scword[chk] += 1
           
        else:
            scword[chk] = 1 
            
       
    elif chk.find("'") != -1:
        chk=chk.replace("'", "")
        if chk in scword:
            scword[chk] += 1
            
        else:
            scword[chk] = 1
            
        
    else:
        nword[chk]= word_dict[chk]  
       
# total num of words from text of with space xx1
xx1=0
for x in word_dict.values():
    xx1 += x

#  total num of words without special characters(,.:;'!?') something like that  xx2
xx2=0
for x in nword.values():
    xx2 += x
    
# total num of words with special characters  xx3
xx3=0
for x in scword.values():
    xx3 += x
    
for x in scword :
    if x in nword:
      nword[x] += scword[x]
    else:
      nword[x] = scword[x]
# total num of words merging words with special character to words without special character xx4=xx2+xx3

xx4=0
for x in nword.values():
    xx4 += x    

    
print(f'Total number of lines in text = {line}')
print(f'Total number of words in text from word_dict={xx1}')
print(f'Total number of words without special chr. in text xx2={xx2}')
print(f'Total number of words with special chr in text xx3={xx3}')
print(f'Total number of words in text xx4={xx4}')
x23=xx2+xx3
print(x23)
x=xx1-xx4
print(x)

print(scword)
print(nword)
print(word_dict)
```
